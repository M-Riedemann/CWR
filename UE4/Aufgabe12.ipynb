```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Einlesen der Daten für den Plot mit Variablen k
kdata = np.loadtxt("position13.csv", delimiter = ",", skiprows = 1)
k = pd.DataFrame(kdata)
edata = np.loadtxt("energy13.csv", delimiter = ",", skiprows = 1)
e = pd.DataFrame(edata)


x_array = k.iloc[0, :].values
t_array = k.iloc[:, 0].values

kin_array = e.iloc[:, 0].values
pot_array = e.iloc[:, 1].values




#Einlesen der Daten für den Plot mit variablen k für verschiedene M



#Plotten 
fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].set_xlabel(r't')
ax[0].set_ylabel(r'Position')
ax[0].grid()
ax[0].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[0].minorticks_on()


for i in range(1,len(x_array)-1):
    
    ax[0].plot(t_array, k.iloc[:, i].values, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-')




ax[1].set_xlabel(r'E_kin')
ax[1].set_ylabel(r'E_pot')
ax[1].grid()
ax[1].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[1].minorticks_on()
ax[1].plot(e.iloc[:,0], e.iloc[:, 1].values, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')


plt.savefig('Positionen.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 
```
