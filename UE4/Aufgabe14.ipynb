```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from mpl_toolkits import mplot3d 
 
```

```python
pos_data = np.loadtxt("A14_pos.csv", delimiter = ",", skiprows = 1)
p = pd.DataFrame(pos_data)


zeit = p.iloc[:, 0].values
x_euler = p.iloc[:, 1].values
y_euler = p.iloc[:, 2].values
x_rk4 = p.iloc[:, 3].values
y_rk4 = p.iloc[:, 4].values


#Plotten 
fig, ax = plt.subplots(1, 2, figsize=(15,5))

ax[0].set_xlabel(r't')
ax[0].set_ylabel(r'$x$')
ax[0].grid()
ax[0].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[0].minorticks_on()
ax[0].plot(zeit, x_euler, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')
ax[0].plot(zeit, x_rk4, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')

ax[1].set_xlabel(r't')
ax[1].set_ylabel(r'$y$')
ax[1].grid()
ax[1].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[1].minorticks_on()
ax[1].plot(zeit, y_euler, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')
ax[1].plot(zeit, y_rk4, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')


plt.savefig('positions.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 

#Plotten 
fig, ax = plt.subplots(1, figsize=(15,5))

ax.set_xlabel(r'x')
ax.set_ylabel(r'$y$')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(x_euler, y_euler, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')
#ax.plot(x_rk4, y_rk4, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')



plt.savefig('positions2.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 

#Plotten 
fig, ax = plt.subplots(1, figsize=(5,5))

ax.set_xlabel(r'x')
ax.set_ylabel(r'$y$')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(zeit, x_euler, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')
ax.plot(zeit, x_rk4, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-')

plt.xlim(1.375,1.435)
plt.ylim(-0.502,-0.495)  

plt.savefig('positions2.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 



```

```python
res_data = np.loadtxt("A14_res.csv", delimiter = ",", skiprows = 1)
p = pd.DataFrame(res_data)

r_zeit = p.iloc[:, 0].values
r_euler = p.iloc[:, 1].values
r_rk4 = p.iloc[:, 2].values

#Plotten 
fig, ax = plt.subplots(1, figsize=(5,5))

ax.set_xlabel(r't')
ax.set_ylabel(r'$residue$')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(r_zeit, r_euler, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-')
ax.plot(r_zeit, r_rk4, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-')



plt.savefig('residue.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 
```
