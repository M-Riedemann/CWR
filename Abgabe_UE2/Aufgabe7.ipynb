```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Einlesen der Daten für den Plot mit Variablen k
kdata = np.loadtxt("data(A07).csv", delimiter = ",", skiprows = 1)
k = pd.DataFrame(kdata)
k_array = k.iloc[:, 0].values
integral_array10 = k.iloc[:, 1].values
integral_array30 = k.iloc[:, 2].values
integral_array100 = k.iloc[:, 3].values
integral_array500 = k.iloc[:, 4].values


#Einlesen der Daten für den Plot mit variablen k für verschiedene M



#Plotten 
fig, ax = plt.subplots(1, 2, figsize=(15,5))
ax[0].set_xlabel(r'k')
ax[0].set_ylabel(r'$\^f(k)$')
ax[0].grid()
ax[0].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[0].minorticks_on()

ax[0].plot(k_array, integral_array10, color = 'red', alpha = 0.75, linewidth = 1, label = r'$M=10$')

ax[0].legend()
ax[0].legend(loc="upper left")

ax[1].set_xlabel(r'k')
ax[1].set_ylabel(r'$\^f_M(k)$')
ax[1].grid()
ax[1].grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax[1].minorticks_on()

ax[1].plot(k_array, integral_array30, color = 'black', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$M=30$')
ax[1].plot(k_array, integral_array100, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$M=100$')
ax[1].plot(k_array, integral_array500, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$M=500$')

ax[1].set_xlim([-4.5,-3.5])
ax[1].set_ylim([-80,550])
ax[1].legend()
ax[1].legend(loc="upper left")


plt.savefig('A07_Fourier.pdf', facecolor = 'white', bbox_inches='tight')
plt.show()
```
