```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Einlesen der Daten für den Plot mit Variablen k
kdata = np.loadtxt("A09_2.0.csv", delimiter = ",", skiprows = 1)
k = pd.DataFrame(kdata)
t_array = k.iloc[:, 0].values
s_array = k.iloc[:, 1].values
e_array = k.iloc[:, 2].values
i_array = k.iloc[:, 3].values
r_array = k.iloc[:, 4].values


#Einlesen der Daten für den Plot mit variablen k für verschiedene M



#Plotten 
fig, ax = plt.subplots()

ax.set_xlabel(r'k')
ax.set_ylabel(r'$\^f_M(k)$')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()

ax.plot(t_array, s_array, color = 'black', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$S$')
ax.plot(t_array, e_array, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$E$')
ax.plot(t_array, i_array, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$I$')
ax.plot(t_array, r_array, color = 'blue', alpha = 0.5, linewidth = 1, linestyle = '-', label = r'$R$')


ax.legend()
ax.legend(loc="upper left")


plt.savefig('Infektionen.pdf', facecolor = 'white', bbox_inches='tight')
plt.show() 
```
