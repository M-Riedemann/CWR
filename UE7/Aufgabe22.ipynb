```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.animation import FuncAnimation
```

```python
#Einlesen der Daten fr den Plot mit Variablen E
Wdata = np.loadtxt("A22_Waermeleitungsgleichung_data.csv", delimiter = ",")
W = pd.DataFrame(Wdata)

x_array = W.iloc[0, 1:].values
t_array = W.iloc[:, 0].values

T_array = W.iloc[1000, 1:].values

print(x_array)
print(t_array)
print(len(T_array))



#Darstellung der Daten
fig, ax = plt.subplots(figsize=(15,5))

ax.set_xlabel(r'T')
ax.set_ylabel(r'x')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(x_array,T_array, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-' )


plt.savefig('A22.pdf', facecolor = 'white', bbox_inches='tight')
plt.show()

# Aufgabe 22 - Wrmeleitungsgleichung
A22_Heat_data = np.loadtxt("A22_Waermeleitungsgleichung_data.csv", delimiter = ",", skiprows = 0)

L = A22_Heat_data[0,1:]
Heat_array = A22_Heat_data[1:2000:10, 1::]
time_array = A22_Heat_data[1:2000:10, 0]


fig, ax = plt.subplots(figsize = (15, 5))

ax.set_xlabel(r'Lnge des Stabes $L$')
ax.set_ylabel(r'Temperatur des Stabes $T$')
ax.grid()
ax.grid(which='minor', color = '#999999', alpha = 0.2, linestyle = '-')
ax.minorticks_on()

for i in range(len(Heat_array)):
    ax.plot(L, Heat_array[i], color = 'red', alpha = time_array[i] / max(time_array), linewidth = 1)#, label = f't = {time_array[i]}')

#ax.legend()
#ax.legend(loc="lower left")

plt.savefig('A22_Waermeleitungsgleichung.pdf', facecolor = 'white', bbox_inches='tight')
plt.show()

```

```python

#Trajektorie. Bitte mit eigenen Daten fllen
trajectory_x = np.linspace(0, 10, 200)
trajectory_y = np.sin(trajectory_x)

fig = plt.figure()
ax = plt.axes()
ax.set_xlim((-0.05,1.05))
ax.set_ylim((-1.1,1.1))
# Linie fr die Trajektorie
line, = ax.plot([], [], lw=3)
# Endpunkt der Trajektorie
point, = ax.plot([], [], lw=3)


def init():
    line.set_data([], [])
    point.set_data([], [])
    return line, point

def animate(i):
    # Funktion, die den Plot auf den Zeitschritt i aktualisiert
    
    # Update line
    x = x_array
    y = A22_Heat_data[i*40][1:]
    line.set_data(x, y)
    
    
    
   
    return line, point

anim = FuncAnimation(fig,
                     animate,
                     init_func=init,
                     frames=200,
                     interval=20,
                     blit=False)

# Speichere als GIF
anim.save('Heat.mp4', writer='imagemagick')
```
