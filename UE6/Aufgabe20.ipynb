```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

 
```

```python
#Einlesen der Daten f√ºr den Plot mit Variablen E
Edata = np.loadtxt("A20_test_energy.csv", delimiter = ",", skiprows = 1)
E = pd.DataFrame(Edata)

test_array = E.iloc[:, 1].values
E_array = E.iloc[:, 0].values
```

```python
#Darstellung der Daten
fig, ax = plt.subplots(figsize=(15,5))

ax.set_xlabel(r'E')
ax.set_ylabel(r'test')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(E_array,test_array, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-' )


plt.savefig('Energie.pdf', facecolor = 'white', bbox_inches='tight')
plt.show()
```

```python
#Transformation der Daten

func = lambda x:  np.sign(x)*np.log(abs(x)+1)

L_test_array = np.zeros(len(test_array))

L_test_array = func(test_array)
```

```python
#Darstellung der Daten
fig, ax = plt.subplots(figsize=(15,5))

ax.set_xlabel(r'E_transformiert')
ax.set_ylabel(r'test')
ax.grid()
ax.grid(which='minor', color = 'grey', alpha = 0.2, linestyle = '-')
ax.minorticks_on()
ax.plot(E_array,L_test_array, color = 'red', alpha = 0.5, linewidth = 1, linestyle = '-' )

plt.savefig('Energie2.pdf', facecolor = 'white', bbox_inches='tight')
plt.show()
```
